```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import (
    train_test_split,
    GridSearchCV,
    StratifiedKFold,
    learning_curve
)
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    roc_auc_score,
    RocCurveDisplay,
    ConfusionMatrixDisplay,
)
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
np.random.seed(42)


diabetes_data = pd.read_csv('diabetes.csv')
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
diabetes_data[zero_cols] = diabetes_data[zero_cols].replace(0, np.nan)
diabetes_data.fillna(diabetes_data.median(), inplace=True)

X_features = diabetes_data.drop('Outcome', axis=1)
y_target = diabetes_data['Outcome']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_features)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_target, test_size=0.2, random_state=42, stratify=y_target
)

model_results = {}
training_histories = {}


def evaluate_model(model_name, model_instance, X_tr, X_te, y_tr, y_te):
    model_history = model_instance.fit(X_tr, y_tr)
    y_pred = model_instance.predict(X_te)
    accuracy = accuracy_score(y_te, y_pred)
    if hasattr(model_instance, "decision_function"):
        y_scores = model_instance.decision_function(X_te)
    else:
        y_scores = model_instance.predict_proba(X_te)[:, 1]
    auc_score = roc_auc_score(y_te, y_scores)
    print(f"\n{model_name} Performance:")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print(f"ROC AUC Score: {auc_score:.4f}")
    model_results[model_name] = {'Accuracy': accuracy, 'ROC AUC': auc_score}
    training_histories[model_name] = model_history
    conf_matrix = confusion_matrix(y_te, y_pred)
    display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
    display.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()
    return y_pred


base_perceptron = Perceptron(
    max_iter=1000,
    eta0=0.01,  
    random_state=42,
    early_stopping=True,
    validation_fraction=0.2,  
    penalty='l2',
    alpha=0.001  
)
y_pred_base = evaluate_model(
    "Base Perceptron Model with L2 Regularization", base_perceptron, X_train, X_test, y_train, y_test
)


perceptron_param_grid = {
    'eta0': [0.001, 0.01],
    'max_iter': [1000, 5000],
    'penalty': ['l2'],
    'alpha': [0.0001, 0.001],
    'early_stopping': [True],
    'validation_fraction': [0.2],
}
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_search_perceptron = GridSearchCV(
    estimator=Perceptron(random_state=42),
    param_grid=perceptron_param_grid,
    cv=cv_strategy,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1,
)
grid_search_perceptron.fit(X_train, y_train)
best_params_perceptron = grid_search_perceptron.best_params_
tuned_perceptron = grid_search_perceptron.best_estimator_
y_pred_tuned = evaluate_model(
    "Tuned Perceptron Model", tuned_perceptron, X_train, X_test, y_train, y_test
)

poly_feat = PolynomialFeatures(degree=2, include_bias=False)
X_poly_features = poly_feat.fit_transform(X_features)
X_poly_scaled = scaler.fit_transform(X_poly_features)
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(
    X_poly_scaled, y_target, test_size=0.2, random_state=42, stratify=y_target
)
poly_perceptron = Perceptron(
    max_iter=1000,
    eta0=0.01,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.2,
    penalty='l2',
    alpha=0.001
)
y_pred_poly = evaluate_model(
    "Perceptron with Polynomial Features and L2 Regularization",
    poly_perceptron,
    X_train_poly,
    X_test_poly,
    y_train_poly,
    y_test_poly,
)

mlp_variations = {
    "MLP 2 Hidden Layers": {
        'hidden_layer_sizes': (100, 50),
        'activation': 'relu',
        'solver': 'adam',
        'alpha': 0.001,
        'learning_rate_init': 0.001  
    },
    "MLP 3 Hidden Layers": {
        'hidden_layer_sizes': (100, 50, 25),
        'activation': 'relu',
        'solver': 'adam',
        'alpha': 0.001,
        'learning_rate_init': 0.001
    },
    "MLP Tanh Activation": {
        'hidden_layer_sizes': (100,),
        'activation': 'tanh',
        'solver': 'adam',
        'alpha': 0.001,
        'learning_rate_init': 0.001
    }
}

for model_name, mlp_params in mlp_variations.items():
    mlp_model_instance = MLPClassifier(
        max_iter=500,
        random_state=42,
        early_stopping=True,
        validation_fraction=0.2,
        **mlp_params,
    )
    y_pred_mlp = evaluate_model(
        model_name, mlp_model_instance, X_train, X_test, y_train, y_test
    )


smote_instance = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote_instance.fit_resample(X_train, y_train)
smote_mlp = MLPClassifier(
    hidden_layer_sizes=(100,),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.2,
    alpha=0.001,
    learning_rate_init=0.001
)

y_pred_smote = evaluate_model(
    "MLP with SMOTE", smote_mlp, X_train_resampled, X_test, y_train_resampled, y_test
)

training_histories["MLP with SMOTE"] = smote_mlp


results_df = pd.DataFrame(model_results).T
results_df.sort_values(by='Accuracy', ascending=False, inplace=True)
print("\nModel Comparison:")
print(results_df)

plt.figure(figsize=(12, 6))
sns.barplot(x=results_df.index, y='Accuracy', data=results_df)
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 8))
models = {
    "Base Perceptron": base_perceptron,
    "Tuned Perceptron": tuned_perceptron,
    "Perceptron with Poly Features": poly_perceptron,
    "MLP 2 Hidden Layers": training_histories["MLP 2 Hidden Layers"],
    "MLP 3 Hidden Layers": training_histories["MLP 3 Hidden Layers"],
    "MLP with SMOTE": training_histories["MLP with SMOTE"],
    "MLP Tanh Activation": training_histories["MLP Tanh Activation"]
}

for model_name, model_instance in models.items():
    if model_name == "Perceptron with Poly Features":
        X_test_final = X_test_poly  
        y_test_final = y_test_poly
    else:
        X_test_final = X_test  
        y_test_final = y_test
    if hasattr(model_instance, "decision_function"):
        y_scores = model_instance.decision_function(X_test_final)
    else:
        y_scores = model_instance.predict_proba(X_test_final)[:, 1]
    RocCurveDisplay.from_predictions(y_test_final, y_scores, name=model_name, ax=plt.gca())

plt.title('ROC Curves for All Models')
plt.show()

plt.figure(figsize=(12, 6))
for model_name in ["MLP 2 Hidden Layers", "MLP 3 Hidden Layers", "MLP Tanh Activation", "MLP with SMOTE"]:
    mlp_instance = training_histories[model_name]
    plt.plot(mlp_instance.loss_curve_, label=model_name)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curves for MLP Models')
plt.legend()
plt.show()


plt.figure(figsize=(12, 6))
train_sizes, train_scores, test_scores = learning_curve(
    MLPClassifier(hidden_layer_sizes=(100, 50, 25), max_iter=500, random_state=42, alpha=0.001),
    X_scaled, y_target, cv=cv_strategy, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5), scoring='accuracy'
)
train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training Score')
plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Cross-validation Score')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.title('Learning Curve for MLP Classifier with 3 Hidden Layers')
plt.legend(loc='best')
plt.show()
```
